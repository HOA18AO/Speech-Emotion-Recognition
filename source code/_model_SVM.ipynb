```python
import _functions
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
from sklearn.metrics import accuracy_score
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
import joblib
```

# PREPROCESSING
label -> dummies

```python
df_TESS = pd.read_csv('TESS_mfcc20_labels.csv')
df_RAVDESS = pd.read_csv('RAVDESS_mfcc20_labels.csv')
```

```python
df = pd.concat([df_TESS, df_RAVDESS], ignore_index=True)
```

```python
df = df.sample(frac=1, random_state=42).reset_index(drop=True) # shuffle
```

```python
df.head(5)
```

```python
labels = df['label'].unique()
labels
```

```python
# Create a LabelEncoder and fit it with the original labels
label_encoder = LabelEncoder()
label_encoder.fit(labels)
```

```python
joblib.dump(label_encoder, 'label_encoder.joblib')
```

```python
df = pd.get_dummies(df, columns=['label'])
```

```python
label_columns = [col for col in df.columns if col.startswith('label_')]
```

```python
X = df.drop(columns=label_columns)
Y = df[label_columns]
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(
    X,
    Y,
    test_size=0.2
)
```

```python
X_train
```

```python
Y_train
```

# MODEL

```python
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score
```

```python
# Convert one-hot encoded labels to single labels for SVM
Y_train_labels = Y_train.idxmax(axis=1)
Y_test_labels = Y_test.idxmax(axis=1)
```

```python
# Initialize the SVM model
svm_model = SVC(kernel='linear', probability=True, random_state=42)
```

```python
svm_model.fit(X_train, Y_train_labels)
```

```python
joblib.dump(svm_model, 'svm_model.joblib')
```

```python
# Predict
Y_pred = svm_model.predict(X_test)
```

```python
# Evaluate
accuracy = accuracy_score(Y_test_labels, Y_pred)
print(f'Accuracy: {accuracy:.2%}')
```
